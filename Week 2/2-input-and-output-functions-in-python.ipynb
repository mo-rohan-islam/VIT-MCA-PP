```python
age = int(input("Enter your age"))
print('Your age is', age)
type(age)
```

```python
cost_of_eggs = 6.5
print(type(cost_of_eggs))   # Not working without print for some reason
print("The cost an egg is \t", cost_of_eggs)
print('The cost of an egg is \n', cost_of_eggs)     # Escape sequence works with both single and double quotes
```

```python
import math

MATH_PI = math.pi
print(MATH_PI)
print(round(MATH_PI, 2))

'''
The following statement allots a total of 
10 positions in the output among which 4 positions 
are being used including the decimal point.

Another method to round off a value
'''
print("%10.2f"%MATH_PI)
print("%.2f"%MATH_PI)       # Only rounds off the value without allotting free space in front

# Other math functions
print(math.pow(5,3))
print(math.sqrt(26))
```
