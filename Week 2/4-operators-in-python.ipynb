##### 6. Special Operators



- ##### Is Operator

The is keyword in Python is used to test object identity.<br/>
The is keyword is used to test whether two variables belong to the same object.<br/>
The test will return True if the two objects are the same else it will return False even if the two objects are 100% equal.

````python
x = ["a", "b", "c", "d"]
y = ["a", "b", "c", "d"]

print('x is not y:', x is not y)
print('x is y:', x is y)
print(x == y)

print('--------------')

x = 10
y = 10

if x is y:
	print(True)
else:
	print(False)
	
print('===============')

# Example 2: Identity check with variables
x = 1000
y = 1000

print(x is y) # Output may vary: False in most cases, because integers above 256 are not always interned

print('----------------')

# Example 3: Using 'is' with singleton objects
n = None
m = None

print(n is m) # Output: True, because there is only one None object in Python, making it a singleton

````

- ##### In Operator

The in keyword in Python returns True if a certain element is present in a Python object, else it will return False. It has two purposes:

- To check if a value is present in a list, tuple, range, string, etc.
- To iterate through a sequence in a for loop.

```python
# Using if statement
if element in sequence:
    # statement

# Using for statement
for element in sequence:
    # statement
```

````python
# Create a list 
animals = ["dog", "lion", "cat"] 

# Check if lion in list or not 
if "lion" in animals: 
	print("Yes")
	
if "tiger" not in animals:
	print("No")

print('=============')

# Create a string 
s = "GeeksforGeeks"

# Iterating through the string 
for i in s: 
	
	# if f occurs in the string 
	# break the loop 
	if i == 'f': 
		break
	
	print(i) 
````
