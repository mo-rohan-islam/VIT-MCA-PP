# Data Types

##### None

```python
x = None # None data type
print(x)
```

##### Numbers

```python
num_int = 5                 # Integer
num_float = 5.57            # Float
num_complex = 4 + 3j        # Complex

print(num_int)
print(num_float)
print(num_complex)
```

##### Binary, Octal and Hexadecimal Literals



```python
# First character is a Zero (0) in each case

num_binary = 0b1010
num_octal = 0o400
num_hexadecimal = 0xED

print(num_binary)
print(num_octal)
print(num_hexadecimal)

# Convert from Decimal to other Bases

num = 50

print(bin(num))
print(oct(num))
print(hex(num))
```

##### Complex Numbers



```python
num_real = 5
num_imaginary = 3

num_complex = complex(num_real, num_imaginary)
print(num_complex)
print(complex(num_real))

print('\nComplex number operations')
print(3+4j*2)
print((3+4j)*2)

complex_1 = 1+2j
complex_2 = 2+4j
complex_sum = complex_1 + complex_2
print(complex_sum)

print('\nProperties of a Complex number')
print(num_complex.real)
print(num_complex.imag + 3)
```
