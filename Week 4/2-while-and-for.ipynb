# While and for

### While else

Else does not get executed on executing break

```python
# Check if a number exists between 1 and 10 else print 'Does not exist'

num = int(input('Enter a number: '))

i = 1

while i<=10:
    if i==num:
        print(f'{num} exists between 1 and 10')
        break
    i += 1
else:
    print(f'{num} does not exist between 1 and 10')
```

### For

Used with sequence or range



```python
for letter in 'Python':
    print('Current Letter: ' + letter)
```

```python
for number in range(1,6):
    # print('Current Number: ' + str(number))
    print(f'Current Number: {number}')
```

### Range



```python
# Print all even numbers until N

N = int(input('Enter a number: '))
for num in range(2,N+1,2):
    print(num)
```
