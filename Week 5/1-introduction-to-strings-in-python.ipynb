# Introduction to Strings in Python



```python
course = 'Python3.0!'

print(course[0])
print(course[5])
print(course[-3])
print(course[-10])
print(course[10])
```

### String Operations

```python
first_name = "Rohan"
last_name = "Islam"

print(first_name*3)
print(len(first_name))
print(first_name+last_name)
```

### String Functions

```python
name = 'Rohan Islam'

print(name.capitalize())        # Note that the 'I' in Islam is changed to lower case 'i'

print(name.find('z'))
print(name.find('m'))
print(name.find(' '))
print(name.find('han'))

print(name.upper())
print(name.lower())
```

```python
email = 'rohan.islam2024e@vitstudent.ac.in'

print(email[12])
print(email[12].isdigit())

print(email[0])
print(email[0].isalpha())

print(email.isalnum())
print('abcd1234'.isalnum())

print('2' in email)
print('1' in email)
print('2024' in email)

print(email.count('a'))
print(email.replace('islam', 'nag'))
print(email.swapcase())
print(email.endswith('.in'))

print(len(email))
print(email.center(50))
```
