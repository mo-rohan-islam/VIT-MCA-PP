# String Slicing





```python
greeting = "Hello World!"

print(greeting[1:4])
print(greeting[7:])
print(greeting[:4])
print(greeting[:-3])
print(greeting[:])
print(greeting[::2])
print(greeting[::-1])
```

```python
print('42' + str(1))
print(42 + int('1'))

print('42' + 1)
```

```python
from random import randint


print(ord('z'))
print(chr(97))

rand_int = randint(0, 1024)
print(rand_int)
print(chr(rand_int))
```

```python
num = 13

num_bin = bin(num)
print(num_bin)
num_bin_str = str(num_bin[2:])
print(num_bin_str)

print(int(num_bin_str, 2))

print(int('456', 8))
print(oct(302))

print(int('FF', 16))
print(hex(255))
print(int('FF', 18))           # For some number system with base 18
print(hex(234598))
print(int('sfh', 16))        # ValueError: invalid literal for int() with base 16: 'sfh'
```
